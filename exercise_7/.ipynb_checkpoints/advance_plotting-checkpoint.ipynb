## Continue working with pandas and matplotlib 

```python
# import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read data
# dataset
fp = 'data\helsinki_test.csv'

# Create dataframe from dataset
data = pd.read_csv(fp,  
                   na_values=['*', '**', '***', '****', '*****', '******'],
                   usecols=['YR--MODAHRMN', 'TEMP', 'MAX', 'MIN'],
                   parse_dates=['YR--MODAHRMN'], index_col='YR--MODAHRMN'
                  )
```

```python
# print number of rows
print(f'Number of rows: {len(data)}')
```

```python
# print first 5 rows
data.head()
```

```python
# rename some columns
new_names = {'TEMP': 'TEMP_F'}
data = data.rename(columns=new_names)

# Check 
data.head()
```

### Prepate the data

```python
print('Number of non data values per column')
print(data.isna().sum()) # print the sum of all the non data values per column 
```

__Let's remove all the Nan values from column 'TEMP_F'__

```python
data.dropna(subset=['TEMP_F'], inplace= True) #  use inplace= True to make change permanent
```

```python
print(f'Number of rows after removing Nan data values: {len(data)}')
```

__Let's convert the temperature in Celsius__

```python
data['TEMP_C'] = (data['TEMP_F']- 32) / 1.8
```

```python
# 

# print sample of new changes
data.head()
```

## Using Subplots

`Subplots` are figures where you have multiple plots in different panels of the same figure

__Extracting temperature per month__
<br>
Let's select dta from May, June, July, August of 2017:


```python
may = data.loc[(data.index >= '201705010000') & (data.index < '201706010000')]
may_temps = may['TEMP_C']
june = data.loc[(data.index >= '201706010000') & (data.index < '201707010000')]
june_temps = june['TEMP_C']
july = data.loc[(data.index >= '201707010000') & (data.index < '201708010000')]
july_temps = july['TEMP_C']
aug = data.loc[data.index >= '201708010000' ]
aug_temps = aug['TEMP_C']
```

```python
ax1 = may_temps.plot()
```

```python
ax2 = june_temps.plot()
```

```python
ax3 = july_temps.plot()
```

```python
ax4 = aug_temps.plot()
```

__Finding data bounds__
<br>
Let's set the y-axis so that the upper limit is the maximum temperature + 5 degres in our data (four months), and the lowest is the minimum temperature -5 degres.

```python
# Find lower limit for y-axis
min_temp = min(may_temps.min(), june_temps.min(), july_temps.min(), aug_temps.min())
min_temp = min_temp -5.0 
min_temp = round(min_temp -5.0, 2) # round the temperature to 2 decimals

# Find upper limit for y-axis 
max_temp = max(may_temps.max(), june_temps.max(), july_temps.max(), aug_temps.max())
max_temp = max_temp + 5.0

# Print y-axis min, max
print(f'Min: {min_temp}, Max: {max_temp}')
```

## Create set of Subplots

Plot all four different plots into same figure we can:
* Create 2x2 panel for visualization using Matplotlib's `subplots()`.
* Specify num. of rows and cols. 
* Specify size of figure with `figsize()`, parameter; `width`  and `height` values (inches) as input.

```python
# Create figure
fig, axs = plt.subplots(nrows= 2, ncols= 2, figsize=(14, 8));


# Plot
axs
```

__We can parse these axis into own variables so it is easier to work with them.__

```python
ax11 = axs[0][0]
ax12 = axs[0][1]
ax21 = axs[1][0]
ax22 = axs[1][1]
```

Now we have four different axis variables for different panels in our figure.
* `c` parameter changes color of the line. list of RGB colors [RGB Colors](https://www.rapidtables.com/web/color/RGB_Color.html)
* `lw` parameter controls width of the line
* `ylim` parameter controls y-axis limits

```python
# set plot line with
line_width = 0.5

# plot data
may_temps.plot(ax= ax11, c='blue', lw= line_width, ylim=[min_temp, max_temp])
june_temps.plot(ax= ax12, c='orange', lw= line_width, ylim=[min_temp, max_temp])
july_temps.plot(ax= ax21, c='green', lw= line_width, ylim=[min_temp, max_temp])
aug_temps.plot(ax= ax22, c='brown', lw= line_width, ylim=[min_temp, max_temp])

# Display figure
fig.tight_layout()
fig
```

```python
# Create a new figure and subplots
fig, axs = plt.subplots(nrows=2, ncols= 2, figsize=(12, 8))

# Rename the axes for ease of use
ax11 = axs[0][0]
ax12 = axs[0][1]
ax21 = axs[1][0]
ax22 = axs[1][1]
```

```python
# Set plot line width 
line_width = 1.5

# Plot data
may_temps.plot(ax= ax11, c='blue', lw= line_width
               , ylim= [min_temp, max_temp], grid=True)
june_temps.plot(ax= ax12, c='orange', lw= line_width
               , ylim= [min_temp, max_temp], grid=True)
july_temps.plot(ax= ax21, c='green', lw= line_width
               , ylim= [min_temp, max_temp], grid=True)
aug_temps.plot(ax= ax22, c='brown', lw= line_width
               , ylim= [min_temp, max_temp], grid=True)

# Set figure title 
fig.suptitle('Summer 2017 Seasonal Temperature')

# Rotate x-axis labels so they don't overlap
plt.setp(ax11.xaxis.get_majorticklabels(), rotation=20)
plt.setp(ax12.xaxis.get_majorticklabels(), rotation=20)
plt.setp(ax21.xaxis.get_majorticklabels(), rotation=20)
plt.setp(ax22.xaxis.get_majorticklabels(), rotation=20)

# Axis label
ax21.set_xlabel('Date')
ax22.set_xlabel('Date')
ax11.set_ylabel('Temperature [C]')
ax21.set_ylabel('Temperature [C]')

# Season label Text
ax11.text(pd.to_datetime('20170528'), -15, 'May')
ax12.text(pd.to_datetime('20170628'), -15, 'June')
ax21.text(pd.to_datetime('20170728'), -15, 'July')
ax22.text(pd.to_datetime('20170828'), -15, 'August')

# Display plot
# fig.tight_layout()
fig

fig.savefig('img/grids')
```

__Let's practice one more time__
<br>
- Visualize May and August temperatures in a 1x2 panel figure.
- Save the figure as a .png file

```python
# Create two subplots side by side
fig, axs = plt.subplots(nrows= 1, ncols= 2, figsize=(12, 4))

# Set plot line width
line_width = 1.5

# Plot data
may_temps.plot(ax=axs[0], c='blue', lw= line_width,
              ylim= [min_temp, max_temp], grid= True)
aug_temps.plot(ax=axs[1], c='green', lw= line_width,
              ylim= [min_temp, max_temp], grid= True)

# Set figure title
fig.suptitle('May and August of 2017 Temperatures observations')

# Rotate the x-axis labels so they don't overlap
plt.setp(axs[0].xaxis.get_majorticklabels(), rotation=20)
plt.setp(axs[1].xaxis.get_majorticklabels(), rotation=20)

# Set axis labels
# axs[0].set_xlabel('Date')
# axs[1].set_xlabel('Date')
# axs[0].set_ylabel('Temperature [C]')
# axs[1].set_ylabel('Temperature [C]')
# OR....
# Set axis labels for all subplots using a For LOOP
for ax in axs.flat:
    ax.set(xlabel='Date', ylabel= 'Temperature [C]')

# Season Label Text
axs[0].text(pd.to_datetime('20170515'), -10, 'May')
axs[1].text(pd.to_datetime('20170815'), -10, 'August')

# Save as .png file
plt.savefig('img/HelsinkiVantaa_MayAugust_2017.png')
```
